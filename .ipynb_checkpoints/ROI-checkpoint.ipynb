```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
!pip install yfinance
import yfinance as yf
%matplotlib inline
!pip install pandas_datareader
import pandas_datareader.data as web


start = "2012-01-01"
end ='2022-12-31'
reliance = yf.download('RELIANCE.NS',start,end).reset_index()
tcs = yf.download('TCS.NS',start,end).reset_index()
hdb = yf.download('HDB',start,end).reset_index()
infy = yf.download('INFY',start,end).reset_index()
ibn = yf.download('IBN',start,end).reset_index()
```

```python
reliance["Date"]=pd.to_datetime(reliance["Date"])
hdb["Date"]=pd.to_datetime(hdb["Date"])
tcs["Date"]=pd.to_datetime(tcs["Date"])
ibn["Date"]=pd.to_datetime(ibn["Date"])
infy["Date"]=pd.to_datetime(infy["Date"])
```

```python
reliance['Year']=reliance["Date"].dt.year
reliance['Day']=reliance["Date"].dt.day
reliance['Month']=reliance["Date"].dt.month

hdb['Month']=hdb["Date"].dt.month
hdb['Year']=hdb["Date"].dt.year
hdb['Day']=hdb["Date"].dt.day

tcs['Day']=tcs['Date'].dt.day
tcs['Year']=tcs['Date'].dt.year
tcs['Month']=tcs['Date'].dt.month

ibn['Day']=ibn['Date'].dt.day
ibn['Year']=ibn['Date'].dt.year
ibn['Month']=ibn['Date'].dt.month

infy['Day']=infy['Date'].dt.day
infy['Year']=infy['Date'].dt.year
infy['Month']=infy['Date'].dt.month
```

```python
#price comparison
plt.figure(figsize=(20,7))

plt.plot(reliance['Date'],reliance['Open'],color='blue',label='Reliance')
plt.plot(hdb['Date'],hdb['Open'],color='grey',label='hdb')
plt.plot(tcs['Date'],tcs['Open'],color='orange',label='TCS')
plt.plot(ibn['Date'],ibn['Open'],color='red',label='IBN')
plt.plot(infy['Date'],infy['Open'],color='yellow',label='IBN')



plt.title("Relation between Reliance, HDB, IBN, Infosys and TCS Price")

plt.xlabel("Year")

plt.ylabel("Price")

plt.legend(title="")

plt.show()
```

```python
#Volume comparison
plt.figure(figsize=(20,7))

plt.plot(reliance['Date'],reliance['Volume'],color='blue',label='Reliance')
plt.plot(hdb['Date'],hdb['Volume'],color='grey',label='hdb')
plt.plot(tcs['Date'],tcs['Volume'],color='orange',label='TCS')
plt.plot(ibn['Date'],ibn['Volume'],color='red',label='IBN')
plt.plot(infy['Date'],infy['Volume'],color='yellow',label='IBN')



plt.title("Relation between Reliance, HDB, IBN, Infosys and TCS Volume")

plt.xlabel("Year")

plt.ylabel("Volume")

plt.legend(title="")

plt.show()
```

```python
sumReliance=0 #total amount invested in Tata Motors
s1=0 #number of shares owned by Tata Motors
#calcuating total amount invested and number of shares owned in Tata Motors
for i in range(len(reliance)):
    if reliance.loc[i,'Day']==30:
        sumReliance+=reliance.loc[i,'Open']
        s1+=1

#displaying basic results
print("Total Invested in Reliance = Rs",round(sumReliance,2))
print("Shares Owned of Reliance =",s1)
print("Average Investmentment of 1 share = Rs",round((sumReliance/s1),2))

r_end=2538.00 #last open price of reliance on 2022-12-28 (today)

#obtained by looking at the data or can be seen after executing tata_motors.tail()
#calculating investment results
result1=round((r_end*s1)-sumReliance,2)
roiReliance=round((result1/sumReliance)*100,2)
#displaying investment results

print("nInvestment Result:")
if result1<0:
    print("Net Unrealised Loss = Rs",result1)
else:
    print("Net Unrealised Profit = Rs",result1)
print("Reliance ROI from 2000-1-3 to 2021-04-30 =",roiReliance,"%")
```

```python
sumhdb=0 #total amount invested in Tata Motors
hdbShares=0 #number of shares owned by Tata Motors
#calcuating total amount invested and number of shares owned in Tata Motors
for i in range(len(hdb)):
    if hdb.loc[i,'Day']==30:
        sumhdb+=hdb.loc[i,'Open']
        hdbShares+=1

#displaying basic results
print("Total Invested in hdb = Rs",round(sumhdb,2))
print("Shares Owned of hdb =",hdbShares)
print("Average Investmentment of 1 share = Rs",round((sumhdb/hdbShares),2))
r_end=67.51 #last open price of tcs on 2022-12-28 (today)

#obtained by looking at the data or can be seen after executing tata_motors.tail()
#calculating investment results
result2=round((r_end*hdbShares)-sumhdb,2)
roihdb=round((result1/sumhdb)*100,2)
#displaying investment results
print("nInvestment Result:")
if result1<0:
    print("Net Unrealised Loss = Rs",result2)
else:
    print("Net Unrealised Profit = Rs",result2)

print("hdb ROI from 2000-1-3 to 2021-04-30 =",roihdb,"%")
```

```python
sumtcs=0 #total amount invested in Tata Motors
tcsShares=0 #number of shares owned by Tata Motors
#calcuating total amount invested and number of shares owned in Tata Motors
for i in range(len(tcs)):
    if tcs.loc[i,'Day']==30:
        sumtcs+=tcs.loc[i,'Open']
        tcsShares+=1

#displaying basic results
print("Total Invested in tcs = Rs",round(sumtcs,2))

print("Shares Owned of tcs =",tcsShares)

print("Average Investmentment of 1 share = Rs",round((sumtcs/tcsShares),2))
r_end=3249.80 #last open price of Tata Motors on 2022-12-28 (today)

#obtained by looking at the data or can be seen after executing tata_motors.tail()
#calculating investment results
result3=round((r_end*tcsShares)-sumtcs,2)
roitcs=round((result3/sumtcs)*100,2)
#displaying investment results
print("nInvestment Result:")
if result3<0:
    print("Net Unrealised Loss = Rs",result3)
else:
    print("Net Unrealised Profit = Rs",result3)

print("tcs ROI from 2000-1-3 to 2021-04-30 =",roitcs,"%")
```
