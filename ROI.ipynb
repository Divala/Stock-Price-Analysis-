```python
reliance["Date"]=pd.to_datetime(reliance["Date"])
hdb["Date"]=pd.to_datetime(hdb["Date"])
tcs["Date"]=pd.to_datetime(tcs["Date"])
```

```python
reliance['Year']=reliance["Date"].dt.year
reliance['Day']=reliance["Date"].dt.day
reliance['Month']=reliance["Date"].dt.month

hdb['Month']=hdb["Date"].dt.month
hdb['Year']=hdb["Date"].dt.year
hdb['Day']=hdb["Date"].dt.day

tcs['Day']=tcs['Date'].dt.day
tcs['Year']=tcs['Date'].dt.year
tcs['Month']=tcs['Date'].dt.month
```

```python
#price comparison
plt.figure(figsize=(20,7))

plt.plot(reliance['Date'],reliance['Open'],color='blue',label='Reliance Industries')
plt.plot(hdb['Date'],hdb['Open'],color='red',label='HDFC Bank')
plt.plot(tcs['Date'],tcs['Open'],color='orange',label='Tata Consultancy Services')

plt.title("Relation between Reliance Industries, HDFC Bank and Tata Consultancy Services by price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend(title="")

plt.show()
```

```python
#Volume comparison
plt.figure(figsize=(20,8))

plt.plot(reliance['Date'],reliance['Volume'],color='blue',label='Reliance Industries')
plt.plot(hdb['Date'],hdb['Volume'],color='red',label='HDFC Bank')
plt.plot(tcs['Date'],tcs['Volume'],color='orange',label='Tata Consultancy Services')

plt.title("Relation between Reliance Industries, HDFC Bank, and Tata Consultancy Services by volume")

plt.xlabel("Year")
plt.ylabel("Volume")
plt.legend(title="")

plt.show()
```

```python
sumReliance=0 #total amount invested in Reliance Industries
relianceShares=0 #number of shares owned by Reliance Industries
#calcuating total amount invested and number of shares owned in Reliance Industries
for i in range(len(reliance)):
    if reliance.loc[i,'Day']==30:
        sumReliance+=reliance.loc[i,'Open']
        relianceShares+=1

#displaying basic results
print("Total Invested in Reliance Industries = $", round(sumReliance,2))
print("Shares Owned of Reliance Industries =", relianceShares)
print("Average Investmentment of 1 share = $", round((sumReliance/relianceShares),2))

r_end=2527.00 #last open price of reliance on 2022-12-31 

#obtained by looking at the data or can be seen after executing reliance.tail()
#calculating investment results
result1=round((r_end*relianceShares)-sumReliance,2)
roiReliance=round((result1/sumReliance)*100,2)
#displaying investment results

print("Investment Result:")
if result1<0:
    print("Net Unrealised Loss = $",result1)
else:
    print("Net Unrealised Profit = $",result1)
print("Reliance Industries ROI from 2012-01-01 to 2022-12-31 =",roiReliance,"%")
```

```python
sumhdb=0 #total amount invested in HDFC Bank
hdbShares=0 #number of shares owned by HDFC Bank
#calcuating total amount invested and number of shares owned in HDFC Bank
for i in range(len(hdb)):
    if hdb.loc[i,'Day']==30:
        sumhdb+=hdb.loc[i,'Open']
        hdbShares+=1

#displaying basic results
print("Total Invested in HDFC Bank = $",round(sumhdb,2))
print("Shares Owned of HDFC Bank =",hdbShares)
print("Average Investmentment of 1 share = $",round((sumhdb/hdbShares),2))
r_end=67.510002 #last open price of tcs on 2022-12-31 

#obtained by looking at the data or can be seen after executing hdb.tail()
#calculating investment results
result2=round((r_end*hdbShares)-sumhdb,2)
roihdb=round((result1/sumhdb)*100,2)
#displaying investment results
print("Investment Result:")
if result1<0:
    print("Net Unrealised Loss = $",result2)
else:
    print("Net Unrealised Profit = $",result2)

print("HDFC Bank ROI from 2012-01-01 to 2022-12-31 =",roihdb,"%")
```

```python
sumtcs=0 #total amount invested in Tata Consultancy Services 
tcsShares=0 #number of shares owned by Tata Consultancy Services 
#calcuating total amount invested and number of shares owned in Tata Consultancy Services 
for i in range(len(tcs)):
    if tcs.loc[i,'Day']==30:
        sumtcs+=tcs.loc[i,'Open']
        tcsShares+=1

#displaying basic results
print("Total Invested in tcs = $",round(sumtcs,2))

print("Shares Owned of tcs =",tcsShares)

print("Average Investmentment of 1 share = $",round((sumtcs/tcsShares),2))
r_end=3231.100098 #last open price of Tata Motors on 2022-12-31 (today)

#obtained by looking at the data or can be seen after executing tcs.tail()
#calculating investment results
result3=round((r_end*tcsShares)-sumtcs,2)
roitcs=round((result3/sumtcs)*100,2)
#displaying investment results
print("Investment Result:")
if result3<0:
    print("Net Unrealised Loss = $",result3)
else:
    print("Net Unrealised Profit = $",result3)

print("tcs ROI from 2012-01-01 to 2022-12-31 =",roitcs,"%")
```
